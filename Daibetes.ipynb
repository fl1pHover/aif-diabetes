# Understand the features



a. Pregnancies - Төрөлт
b. Glucose - Глюкоз ( Амны хөндийн глюкозын хүлцэл. 2 цагийн шинжилгээний сийвэн дэх глюкозын концентраци )
c. Blood Pressure - Цусны даралт (mm Hg) ( <120/<80 , 130/85 байх нь бага зэрэг өндөр гэж үзнэ. 140/90 дээш байвал ямар нэг өвчин зүдгүүртэй )
d. Skin Thickness - Арьсны зузаан (mm)
e. Insulin - 2 цагийн сийвэнгийн инсулин даавар (мУ/мл) 
f. BMI - Body Mass Index - Биеийн жингийн индекс (weight in kg/(height in m)^2) ( Эрүүл биеийн жингийн түвшинг илэрхийлэх, биеийн жинг биеийн өндрийн квадрат зэрэгт дэвшүүлсэнтэй харьцуулж тооцдог үзүүлэлт )
g. Diabetes Pedigree Function - Чихрийн шижингийн удамшлын функц ( Гэр бүлийн түүх дээр үндэслэн чихрийн шижин өвчний магадлалыг оноодог функц )
h. Age - Нас
i. Outcome - Анги давхарга (0 or 1) ( 268 of 768 are 1, the others are 0 )

```python
dt = diabetes

def Mode(x):
    a = pd.Series(x).value_counts()
    return a.idxmax()

mode = Mode(dt['Pregnancies'])

mode
```
